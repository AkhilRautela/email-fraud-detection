````python
x_train = data['review'].values[:10000]
y_train = data['sentiment'].values[:10000]
````

````python
x_train = np.array([x[:32] for x in x_train])
````

````python
y_train = np.array([0 if x == "positive" else 1 for x in y_train])
````

````python
y_train
````

````python
import re
def clean_text(text):
  if type(text)!=str:
    return str(text)
  return re.sub(r'[^0-9A-Za-z" "]','',text)
````

````python
x_train = [clean_text(here) for here in x_train]
save =[]
for each_sentence in x_train:
  save.append(each_sentence.split(" "))
    
````

````python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(x_train)
````

````python
x_train = tokenizer.texts_to_matrix(x_train, mode='count')
````

````python
model = LogisticRegression()
````

````python
x_train, x_test, y_train, y_test = train_test_split(x_train , y_train , test_size=0.2, random_state=42)
````

````python
model.fit(x_train, y_train)
````

````python
y_pred = model.predict(x_test)
````

````python
confusion_matrix(y_test, y_pred)
````

````python
print ("Accuracy : ", accuracy_score(y_test, y_pred))
````
